```python
import xml.etree.cElementTree as et
import pandas as pd
 
def getnodevalue(node):
    """ return node  or None """
    return node if node is not None else None

def main():
    """ main """
    parsed_xml = et.parse("activationData.xml")
    dfxmlcols = ['EventId', 'InvestigationType', 'AccidentNumber', 'EventDate', 'Location', 
    'Country', 'Latitude', 'Longitude'
]
    dffromxml = pd.DataFrame(columns=dfxmlcols)
 
    for a in parsed_xml.getroot():
        for node in a:
            ID = node.attrib.get('EventId')
            type = node.attrib.get('InvestigationType')
            Anumber = node.attrib.get('AccidentNumber')
            Edate = node.attrib.get('EventDate')
            Location = node.attrib.get('Location')
            type = node.attrib.get('InvestigationType')
            Country = node.attrib.get('Country')
            Latitude = node.attrib.get('Latitude')
            dffromxml = dffromxml.append(
                pd.Series([ID, getnodevalue(type), getnodevalue(Anumber),
                       getnodevalue(Edate), getnodevalue(Location),getnodevalue(type),getnodevalue(Country),
                       getnodevalue(Latitude) ], index=dfxmlcols),
            ignore_index=True)
    print (dffromxml.head())
main()
```
